###VARIANZA###


```python
lista = [0, 1, 2, 3, 4, 5, 6, 7, 8]

if len(lista) != 9:
    print("La lista DEBE CONTENER 9 elementos.")
else:
    print("La lista tiene 9 elementos.")
```

```python
import numpy as np

arreglo = np.array(lista)
print(arreglo)
#Convirtiendo la lista an Numpy
```

# 3 * **3** ‚õπ

```python
arreglo = arreglo.reshape(3, 3)
print(arreglo)
```

```python
 # Convertir la lista en un arreglo de Numpy 3x3
arreglo = np.array(lista).reshape(3, 3)

# Calcular las estad√≠sticas
media = [np.mean(arreglo, axis=0).tolist(), np.mean(arreglo, axis=1).tolist(), np.mean(arreglo).tolist()]
varianza = [np.var(arreglo, axis=0).tolist(), np.var(arreglo, axis=1).tolist(), np.var(arreglo).tolist()]
desviacion_std = [np.std(arreglo, axis=0).tolist(), np.std(arreglo, axis=1).tolist(), np.std(arreglo).tolist()]
maximo = [np.max(arreglo, axis=0).tolist(), np.max(arreglo, axis=1).tolist(), np.max(arreglo).tolist()]
minimo = [np.min(arreglo, axis=0).tolist(), np.min(arreglo, axis=1).tolist(), np.min(arreglo).tolist()]
suma = [np.sum(arreglo, axis=0).tolist(), np.sum(arreglo, axis=1).tolist(), np.sum(arreglo).tolist()]

# Crear el diccionario con los resultados
resultados = {
    'media': media,
    'varianza': varianza,
    'desviacion_std': desviacion_std,
    'maximo': maximo,
    'minimo': minimo,
    'suma': suma
   }
print(resultados)
```

```python
import pandas as pd

# Modify the 'media' list in 'resultados' to have consistent dimensions
resultados['media'][2] = [resultados['media'][2]]  # Enclose the single float value in a list
resultados['varianza'][2] = [resultados['varianza'][2]]  # Enclose the single float value in a list
resultados['desviacion_std'][2] = [resultados['desviacion_std'][2]]
resultados['maximo'][2] = [resultados['maximo'][2]]
resultados['minimo'][2] = [resultados['minimo'][2]]
resultados['suma'][2] = [resultados['suma'][2]]

# Create a DataFrame for each metric
df_media = pd.DataFrame(resultados['media'], index=['Columnas', 'Filas', 'Total'], columns=['Col1', 'Col2', 'Col3'])
df_varianza = pd.DataFrame(resultados['varianza'], index=['Columnas', 'Filas', 'Total'], columns=['Col1', 'Col2', 'Col3'])
df_desviacion_std = pd.DataFrame(resultados['desviacion_std'], index=['Columnas', 'Filas', 'Total'], columns=['Col1', 'Col2', 'Col3'])
df_maximo = pd.DataFrame(resultados['maximo'], index=['Columnas', 'Filas', 'Total'], columns=['Col1', 'Col2', 'Col3'])
df_minimo = pd.DataFrame(resultados['minimo'], index=['Columnas', 'Filas', 'Total'], columns=['Col1', 'Col2', 'Col3'])
df_suma = pd.DataFrame(resultados['suma'], index=['Columnas', 'Filas', 'Total'], columns=['Col1', 'Col2', 'Col3'])

# Mostrar los DataFrames
print("Media:\n", df_media)
print("\nVarianza:\n", df_varianza)
print("\nDesviaci√≥n Est√°ndar:\n", df_desviacion_std)
print("\nM√°ximo:\n", df_maximo)
print("\nM√≠nimo:\n", df_minimo)
print("\nSuma:\n", df_suma)
```

```python
import matplotlib.pyplot as plt

# Create a single figure and axes
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figsize as needed

# Plot the 'Media' data on the single axes
ax.bar(['Col1', 'Col2', 'Col3'], resultados['media'][0], label='Columnas')
ax.bar(['Fila1', 'Fila2', 'Fila3'], resultados['media'][1], label='Filas', alpha=0.7)

# Check if resultados['media'][2][0] is None and replace with 0 if necessary
total_media = resultados['media'][2][0] if resultados['media'][2][0] is not None else 0

# Set title and legend
ax.set_title('Media')
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
